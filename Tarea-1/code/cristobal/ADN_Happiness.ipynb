```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go


df = pd.read_csv('songs.csv', nrows=499000, encoding='latin1')

metricas = ['valence', 'energy', 'acousticness', 'danceability', 'speechiness'] 
# valencia es alegría general por lo que entiendo
# energía es, bueno, energía.
# las otras son intuibles.

df['Decada'] = (df['year'] // 10) * 10 # esto pa formatear todas las canciones y que se clasifiquen por década
df_comp = df[df['Decada'].isin([1970, 1980, 1990, 2010, 2020])].copy() # cada decada por separado
df_comp['Decada_str'] = df_comp['Decada'].astype(int).astype(str) + "s" # ese "s" es para agregarle un 2000 ->s<- ==> 2000s the twothousandSSS. 

df_radar = df_comp.groupby('Decada_str')[metricas].mean().reset_index()

df_radar_long = df_radar.melt(id_vars=['Decada_str'], var_name='Metrica', value_name='Valor')

fig = px.line_polar( # visualización que espero sea lo suficientemente única
    df_radar_long, 
    r='Valor', 
    theta='Metrica', 
    color='Decada_str',
    line_close=True,
    template="plotly_dark", 
    title="ADN Emocional de la Música durante las decadas",
    color_discrete_sequence=['#00CC96',"#02F31E","#ECD60C","#EC6D0C", "#E63C1D"] # azulito a rojo, antes eramos más felices :(
)

fig.update_traces(
    fill='toself', 
    opacity=0.4,      # poco menos de opacida
    line=dict(width=3) # mas grueso para destacar :)
)

fig.show()

```
